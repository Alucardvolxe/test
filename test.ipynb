```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr = np.array([[-2,-1],[2,-1],[2,1],[-2,1]])
```

```python
rectangle_closed = np.vstack([arr, arr[0]])
plt.figure(figsize=(5,5))
plt.plot(rectangle_closed[:,0], rectangle_closed[:,1], 'b-', linewidth =2)
plt.scatter(rectangle_closed[:,0],rectangle_closed[:,1],color='red')
plt.title("Rectangle")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.axis("equal") 
plt.grid(True)
plt.fill(
    rectangle_closed[:,0],
    rectangle_closed[:,1],
    color = 'red',
    alpha =0.6
)

plt.show()
```
