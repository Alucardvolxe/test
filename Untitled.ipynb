```python
def factorial(n):
    if(n<=1):
        return 1
    
    
    return n*factorial(n-1)


print(factorial(3))
```

```python
def sub(n):
    if (n<=2):
        return 0
    print(n)
    return sub(n-2)+sub(n-1)

print(sub(6))
```

```python
def revers_word(word):
    if len(word)==0:
        return ""
    else:
        return revers_word(word[1:])+word[0]
    
revers_word("hello")
```
