```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr = np.array([[-2,-1],[2,-1],[2,1],[-2,1]])
```

```python
rectangle_closed = np.vstack([arr, arr[0]])
plt.figure(figsize=(5,5))
plt.plot(rectangle_closed[:,0], rectangle_closed[:,1], 'b-', linewidth =2)
plt.scatter(rectangle_closed[:,0],rectangle_closed[:,1],color='red')
plt.title("Rectangle")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.axis("equal") 
plt.grid(True)
plt.fill(
    rectangle_closed[:,0],
    rectangle_closed[:,1],
    color = 'red',
    alpha =0.6
)

plt.show()
```

```python
# Scaling matrix (2x zoom)
scale_matrix = np.array([
    [3, 0],
    [0, 3]
])

# Original coordinates
points = np.array([
    [0, 0],
    [1, 0],
    [1, 1],
    [0, 1]
])

scaled_points = points @ scale_matrix

plt.plot(points[:,0], points[:,1], 'o-', label='Original')
plt.plot(scaled_points[:,0], scaled_points[:,1], 'o-', label='Scaled')
plt.legend()
plt.title("Scaling using Matrix Multiplication")
plt.axis('equal')
plt.show()
```

```python
# Rotation matrix (90 degrees)
rotation_matrix = np.array([
    [0, -2],
    [2,  0]
])

rotated_points = points @ rotation_matrix

plt.plot(points[:,0], points[:,1], 'o-', label='Original')
plt.plot(rotated_points[:,0], rotated_points[:,1], 'o-', label='Rotated')
plt.legend()
plt.title("Rotation using a Matrix")
plt.axis('equal')
plt.show()
```

```python
# Rotation matrix (90 degrees)
rotation_matrix = np.array([
    [0, -1],
    [1,  0]
])

rotated_points = points @ rotation_matrix

plt.plot(points[:,0], points[:,1], 'o-', label='Original')
plt.plot(rotated_points[:,0], rotated_points[:,1], 'o-', label='Rotated')
plt.legend()
plt.title("Rotation using a Matrix")
plt.axis('equal')
plt.show()
```

```python
flip_matrix = np.array([
    [-1, 0],
    [ 0, 1]
])

flipped_points = points @ flip_matrix

plt.plot(points[:,0], points[:,1], 'o-', label='Original')
plt.plot(flipped_points[:,0], flipped_points[:,1], 'o-', label='Flipped')
plt.legend()
plt.title("Image Reflection using a Matrix")
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d

# Original image (random for demo)
image = np.random.rand(10, 10)

# Blur kernel
blur_kernel = np.ones((3, 3)) / 9

# Sharpen kernel
sharpen_kernel = np.array([
    [ 0, -1,  0],
    [-1,  5, -1],
    [ 0, -1,  0]
])
# Apply filters
blurred = convolve2d(image, blur_kernel, mode='same')
sharpened = convolve2d(image, sharpen_kernel, mode='same')

# Display results
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(blurred, cmap='gray')
plt.title("Blurred Image")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(sharpened, cmap='gray')
plt.title("Sharpened Image")
plt.axis('off')

plt.show()


import numpy as np
import matplotlib.pyplot as plt

# Define a square (image outline)
points = np.array([
    [0, 0],
    [1, 0],
    [1, 1],
    [0, 1],
    [0, 0]   # close the shape
])

# Scaling matrix (2x in x and y)
scaling_matrix = np.array([
    [2, 0],
    [0, 2]
])

# Apply scaling
scaled_points = points @ scaling_matrix

# Plot
plt.figure()
plt.plot(points[:, 0], points[:, 1], marker='o', label='Original')
plt.plot(scaled_points[:, 0], scaled_points[:, 1], marker='o', label='Scaled (2x)')
plt.legend()
plt.title("Scaling using a Matrix")
plt.axis('equal')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a 10x10 RGB image
image_rgb = np.zeros((10, 10, 3))

# Red channel
image_rgb[:, :, 0] = 1.0   # full red

# Green channel
image_rgb[3:7, 3:7, 1] = 1.0

# Blue channel
image_rgb[5:9, 5:9, 2] = 1.0

plt.imshow(image_rgb)
plt.title("RGB Image as Three Matrices")
plt.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a sample grayscale image
image = np.random.rand(50, 50)

# Increase brightness (scalar multiplication)
bright_image = np.clip(image * 1.5, 0, 1)

# Decrease brightness
dim_image = np.clip(image * 0.5, 0, 1)

# Display results
plt.figure(figsize=(9, 3))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(bright_image, cmap='gray')
plt.title("Brightened")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(dim_image, cmap='gray')
plt.title("Dimmed")
plt.axis('off')

plt.show()
```
