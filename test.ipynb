```python
# Scaling matrix (2x zoom)
scale_matrix = np.array([
    [3, 0],
    [0, 3]
])

# Original coordinates
points = np.array([
    [0, 0],
    [1, 0],
    [1, 1],
    [0, 1]
])

scaled_points = points @ scale_matrix

plt.plot(points[:,0], points[:,1], 'o-', label='Original')
plt.plot(scaled_points[:,0], scaled_points[:,1], 'o-', label='Scaled')
plt.legend()
plt.title("Scaling using Matrix Multiplication")
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d

# Original image (random for demo)
image = np.random.rand(10, 10)

# Blur kernel
blur_kernel = np.ones((3, 3)) / 9

# Sharpen kernel
sharpen_kernel = np.array([
    [ 0, -1,  0],
    [-1,  5, -1],
    [ 0, -1,  0]
])
# Apply filters
blurred = convolve2d(image, blur_kernel, mode='same')
sharpened = convolve2d(image, sharpen_kernel, mode='same')

# Display results
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(blurred, cmap='gray')
plt.title("Blurred Image")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(sharpened, cmap='gray')
plt.title("Sharpened Image")
plt.axis('off')

plt.show()


import numpy as np
import matplotlib.pyplot as plt

# Define a square (image outline)
points = np.array([
    [0, 0],
    [1, 0],
    [1, 1],
    [0, 1],
    [0, 0]   # close the shape
])

# Scaling matrix (2x in x and y)
scaling_matrix = np.array([
    [2, 0],
    [0, 2]
])

# Apply scaling
scaled_points = points @ scaling_matrix

# Plot
plt.figure()
plt.plot(points[:, 0], points[:, 1], marker='o', label='Original')
plt.plot(scaled_points[:, 0], scaled_points[:, 1], marker='o', label='Scaled (2x)')
plt.legend()
plt.title("Scaling using a Matrix")
plt.axis('equal')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------
# Parameters
# -------------------------
scale = 3
graph_scale =scale

# -------------------------
# Get coordinates of non-zero pixels
# -------------------------
coords = np.column_stack(np.where(img > 0))

# Image center
center = np.array([img.shape[0] // 2, img.shape[1] // 2])

# -------------------------
# Scaling matrix
# -------------------------
scaling_matrix = np.array([
    [scale, 0],
    [0, scale]
])

# -------------------------
# Center  scale  uncenter
# -------------------------
scaled_coords = (coords - center) @ scaling_matrix.T + center * scale
scaled_coords = scaled_coords.astype(int)

# -------------------------
# Create new scaled image
# -------------------------
new_size = img.shape[0] * scale
scaled_img = np.zeros((new_size, new_size))

# -------------------------
# Nearest-neighbor block filling
# -------------------------
for (r, c), (sr, sc) in zip(coords, scaled_coords):
    if 0 <= sr < new_size - scale and 0 <= sc < new_size - scale:
        scaled_img[
            sr:sr + scale,
            sc:sc + scale
        ] = img[r, c]

# -------------------------
# Display (separate figures to show size difference)
# -------------------------
plt.figure(figsize=(4, 4))
plt.imshow(img, cmap='gray')
plt.title("Original Smiley (100100)")
plt.axis('off')
plt.show()

plt.figure(figsize=(4*graph_scale, 4*graph_scale))
plt.imshow(scaled_img, cmap='gray')
plt.title("Scaled Smiley (2  200200)")
plt.axis('off')
plt.show()



```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d

# -------------------------
# Step 1: Create Smiley (100x100)
# -------------------------
img = np.zeros((100, 100))

# Face outline
rr, cc = np.ogrid[:100, :100]
center_r, center_c = 50, 50
radius = 35
mask = (rr - center_r)**2 + (cc - center_c)**2 <= radius**2
img[mask] = 0.8

# Eyes
img[40:45, 35:40] = 0
img[40:45, 60:65] = 0

# Smile
img[60, 35:65] = 0
img[59, 36] = 0
img[59, 64] = 0

# -------------------------
# Step 2: Blur
# -------------------------
blur_kernel = np.ones((3, 3)) / 9
blurred = convolve2d(img, blur_kernel, mode='same')

# -------------------------
# Step 3: Sharpen using Unsharp Masking (original image sharpened)
# -------------------------
k = 1.5  # sharpening strength
sharpened = img + k * (img - blurred)
sharpened = np.clip(sharpened, 0, 1)

# -------------------------
# Step 4: High-res scaling (2x) with nearest-neighbor block filling
# -------------------------
scale = 2
coords = np.column_stack(np.where(sharpened > 0))  # use sharpened as input
center = np.array([img.shape[0] // 2, img.shape[1] // 2])
scaling_matrix = np.array([[scale, 0], [0, scale]])
scaled_coords = (coords - center) @ scaling_matrix.T + center * scale
scaled_coords = scaled_coords.astype(int)

new_size = img.shape[0] * scale
scaled_img = np.zeros((new_size, new_size))

# Nearest-neighbor block filling
for (r, c), (sr, sc) in zip(coords, scaled_coords):
    if 0 <= sr < new_size - scale and 0 <= sc < new_size - scale:
        scaled_img[sr:sr + scale, sc:sc + scale] = sharpened[r, c]

# -------------------------
# Step 5: Display results
# -------------------------
plt.figure(figsize=(16, 4))

plt.subplot(1, 4, 1)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(blurred, cmap='gray')
plt.title("Blurred")
plt.axis('off')


plt.subplot(1, 4, 3)
plt.imshow(img, cmap='gray')
plt.title("sharpened")
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import rotate

# -------------------------
# Step 1: Create solid smiley (100x100)
# -------------------------
img = np.zeros((100, 100))
rr, cc = np.ogrid[:100, :100]
center_r, center_c = 50, 50
radius = 35
mask = (rr - center_r)**2 + (cc - center_c)**2 <= radius**2
img[mask] = 0.8  # solid face

# Eyes
img[40:45, 35:40] = 0  # left eye
img[40:45, 60:65] = 0  # right eye

# Smile (make a thin line)
img[60, 35:65] = 0
img[59, 36] = 0
img[59, 64] = 0

# -------------------------
# Step 2: Rotate using scipy.ndimage.rotate
# -------------------------
angle_deg = 60
rotated_img = rotate(img, angle_deg, reshape=False, order=1, mode='nearest')

# -------------------------
# Step 3: Display
# -------------------------
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("Original Smiley")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(rotated_img, cmap='gray')
plt.title(f"Rotated Smiley ({angle_deg})")
plt.axis('off')

plt.tight_layout()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a sample grayscale image
image = np.random.rand(50, 50)

# Increase brightness (scalar multiplication)
bright_image = np.clip(image * 1.5, 0, 1)

# Decrease brightness
dim_image = np.clip(image * 0.5, 0, 1)

# Display results
plt.figure(figsize=(9, 3))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(bright_image, cmap='gray')
plt.title("Brightened")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(dim_image, cmap='gray')
plt.title("Dimmed")
plt.axis('off')

plt.show()
```
